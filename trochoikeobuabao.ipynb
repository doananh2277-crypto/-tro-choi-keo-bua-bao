```python
import random
options=['kéo', 'búa', 'bao']
print("----CHÀO MỪNG ĐẾN VỚI GAME KÉO BÚA BAO----")
user_choice= input("Nhập lựa chọn của bạn:").lower()
computer_choice=random.choice(options)
if user_choice == computer_choice:
  print(" HÒA RỒI! BẮT TAY CÁI NÀO.🙌")
elif (user_choice == "kéo" and computer_choice == "bao") or \
     (user_choice == "búa" and computer_choice == "kéo") or \
     (user_choice == "bao" and computer_choice == "búa"):
  print("CHÚC MỪNG BẠN THẮNG RỒI,ĐỈNH QUÁ!")
else:
  print("MÁY TÍNH THẮNG MẤT RỒI, THỬ LẠI VÁN NỮA KHÔNG?")
```
